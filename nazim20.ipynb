```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

### plotting

```python
sns.kdeplot(data=df.PassengerId)
plt.show()
```

```python
sns.histplot(df.Fare)
plt.show()
```

```python
sns.boxplot('Pclass',data=df)
plt.show()
```

```python
sns.lineplot('Age','Fare',data=df)
plt.title('age vs fare')
plt.show()
```

```python
pclass_sex_survived=df.groupby(['Pclass','Sex'])['Survived'].sum()

pclass_sex_survived.reset_index()
```

```python
sns.set_style('dark')
plt.title('pclass_sex_survived_pie chart')
pclass_sex_survived.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

```python
two=df.groupby(['Pclass','Sex'])['Fare'].sum()
two.reset_index()
```

```python

two.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

```python
three=df.corr()
sns.heatmap(three)
plt.show()
```

```python
sns.set_style('ticks')
plt.title('pclass fare bar chart')
sns.barplot('Age','Sex',data=df)                              
```

```python
sns.violinplot('Pclass','Age',hue="Survived",data=df)
sns.set_style('ticks')
plt.title("pclass age violinplot")
```

```python
sns.factorplot('Pclass','Age', hue='Survived',data=df)
sns.set_style('dark')
plt.title('factor plot')
```

```python
sns.kdeplot('Pclass',data=df)
plt.title('countplot')
sns.set_style('darkgrid')
plt.show()
```
